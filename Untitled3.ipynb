```python
import pandas as pd
import numpy as np

# Step 6: Create a table for visualization
nodes = [f'N{i+1}' for i in range(len(dataset))]

# Ensure rank is a NumPy array
rank = np.array(rank)

# --- Handle shape differences safely ---
if rank.ndim == 1:  # rank is 1D → directly MAUT scores
    maut_scores = rank
elif rank.ndim == 2:
    # Check which column looks like scores (choose numeric, not index)
    if rank.shape[1] == 1:
        maut_scores = rank[:, 0]
    else:
        # Automatically pick the column with max variance (likely scores)
        col_var = np.var(rank, axis=0)
        maut_scores = rank[:, np.argmax(col_var)]
else:
    raise ValueError("Unexpected rank array shape!")

# --- Make sure lengths match ---
if len(maut_scores) != len(nodes):
    raise ValueError("Length mismatch between dataset and MAUT results!")

# --- Create final table ---
results = pd.DataFrame({
    'Node': nodes,
    'MAUT_Score': maut_scores
})

# Sort descending (best = top)
results = results.sort_values(by='MAUT_Score', ascending=False).reset_index(drop=True)

# --- Display clean table ---
print(results[['Node', 'MAUT_Score']])

# --- Optional: Print the best node (Cluster Head) ---
print("\nBest Node (Cluster Head):", results.iloc[0]['Node'])
```
