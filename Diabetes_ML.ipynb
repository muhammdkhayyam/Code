```python
from sklearn.datasets import load_diabetes
```

```python
d = load_diabetes()
X,y=load_diabetes(return_X_y=True)
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python

X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.20)
print(X_train)
print("******************************************************************************************************")
print(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python

obj=LinearRegression()
mp=obj.fit(X_train,y_train)
```

```python
print(mp.coef_)
print("*************************************************************************************************")
print(mp.intercept_)
```

```python
predict = mp.score(X_test,y_test)
print("The test part is predict is: ",predict)
```

```python
predict = mp.score(X_train,y_train)
print("The train part is predict is: ",predict)
print("**************************************************************************************")
print("Predict value of X_train")
print(obj.predict(X_train))
```
